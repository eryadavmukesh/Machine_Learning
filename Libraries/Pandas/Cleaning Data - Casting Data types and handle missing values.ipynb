```python
# Imorting required libraries

import pandas as pd
import numpy as np
```

```python
persons={
    'first':['Mukesh','Shubham','Joe','Wille', np.nan, None, 'NA'],
    'last':['Yadav','Sharma','Knack','Meeks',np.nan, np.nan,'Missing'],
    'email':['mukesh@gmail.com','shubham@gmail.com','Joe@live.com',None, np.nan, 'NoName@yahoo.com','NA'],
    'age':['28','25','54','34',None, None, 'Missing']
}
```

```python
# Create a data frame

data = pd.DataFrame(persons)
data
```

```python
# Let's go through our persons data and drop the null rows

data.dropna()
```

```python
# dropna uses defualt parameters 

#You can specify axis='index' (instead of axis=0) or axis='columns' (instead of axis=1).

data.dropna(axis='index', how='any')
```

```python
# if all the rows are satisfying the condition then only it will remove 

data.dropna(axis='index', how='all')

# just check the original dataset and you'll get to know that index 4 has all the rows missing 
```

```python
# we can pass the argument as subset to gain more insight about the data

data.dropna(axis='index', how='any', subset=['email'])
```

```python
data.dropna(axis='index', how='all', subset=['last','email'])

# here last lame at index 3 is there but email is not there, and at index 5 last name is not there but but emial exists
```

```python
# Here at index 6 i have some custom missing values, So we can replace these values as proper numpy NaN values-

# Let's replace all the missing

data = pd.DataFrame(persons)

data.replace('NA', np.nan, inplace=True)
data.replace('Missing', np.nan, inplace=True)

data

# So here we no longer have custom missing values like --> Missing 
```

```python
# Now let's again do dropna

data.dropna()

# Now it will remove all the rows which has missing values
```

```python
data.dropna(axis='index', how='all', subset=['last','email'])
```

```python
# Let's use isna method

data.isna()

# True --> missing values
```

```python
# we can use fillna() to fill null values.
```

```python
data.fillna('MukeshTecHub')

# if data is numeric then we can pass 0 to fill at the place of nulls, like below
```

```python
data.fillna(0)
```

### Now let's look at the casting

```python
# so we have a column named as age and that is type of string- let's check the data types.

data.dtypes
```

```python
# so I can see that age type is Object and that means this is type of string , so if I want to calculate average for the 
#age column then it wouldn't work

data['age'].mean()
```

```python
# so our age colum is string type and for average we required number, SO if a column has NaN and we trying to convert it to number-
# then we need to use float data type, that's bcz nan is a float, to be more clear let me check the type of NaN

type(np.NaN)
```

```python
# so here we can't convert it here into integers, let's try to convert into integers-

data['age'] = data['age'].astype(int)
```

```python
# So data['age'] = data['age'].astype(int) it will work file if ther's no missing values in age column
```

```python

data['age'] = data['age'].astype(float)
```

```python
data.dtypes
```

```python
# Now let's try to take average of that column - 

data['age'].mean()
```
