```python
# I need to make this coulmn as a date type -

df = pd.read_csv('MarriageData.csv', parse_dates=['day'])
```

```python
type(df.day[0])
```

```python
# Now I want to set day column as an index
df.set_index('day', inplace=True)
```

```python
df
```

```python
# Now we can fill the NaN - we can have any values, lets try with 0

df.fillna(0)
```

```python
# If you don;t want to fill all the coulumns with 0, then you can specify the values for columns

modifiedData = df.fillna({
    'Couples Got Married':0,
    'Total Arriavls':0,
    'Gifts Collected':'No Gifts As Such'
    
})
modifiedData

# So below 2 columns got replaces with zeros and last with No Gifts As Such
```

```python
# SO here still we can think a better approach for filling the rows instead of 0, even we can fill the previous row value

# we will use forward fill method it will fill the previous values, to fill next day values we can use bfill method

modifiedData = data.fillna(method='ffill')
modifiedData
```

```python
modifiedData
```

```python
# SO if you want to fill the values upto level 1 , i mean only for next 1 row then we can use limit

modifiedData = data.fillna(method='ffill', limit=1)
modifiedData
```

```python
# Instead of filling back rows data we can use interpolate method

modifiedData = data.interpolate()
modifiedData
```

```python
# Also we can use dropna

modifiedData = data.dropna()
modifiedData

# Here it will delete all the rows which contians NaN values
```
