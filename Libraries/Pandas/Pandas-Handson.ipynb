### Sorting of the data


```python
# Sort the data based on the contry column

df.sort_values(by='Country', inplace=True)
df
```

```python
# Here I can see that data is sorted

df['Country'].head(50)
```

```python
# Always use another list for multiple list of columns. 

df[['Country', 'SalaryUSD']].head(50)
```

```python
#Here I can see that many rows has NaN, if you want to have max salary first-

df.sort_values(by=['Country','SalaryUSD'], ascending=[True, False], inplace=True)
```

```python
# Now run again the code for contry and salaryUSD and can seee the difference

df[['Country', 'SalaryUSD']].head(50)
```

```python
# Note --> so if you can observe we have a problem of outliers in our data, 62723	Afghanistan	1000000.0 this has a 
# max notable difference from other records, so this is called outlier problem.
```

```python
# Now find the 15 largest salary records 

df['SalaryUSD'].nlargest(15)
```

```python
# so instead of just salary if you want to retrieve all the columns along with 15 max salary -

df.nlargest(15, ['SalaryUSD'])
```
