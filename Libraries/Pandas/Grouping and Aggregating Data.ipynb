### GroupBy function
- groupby() function is used to split the data into groups based on some criteria.
- it splits the data , applies a functiona and combines the results.

```python
# If we want to see specific results bases on the country or based on some other columns then going to have group for that specific column.

dataset['Country'].value_counts()
```

```python
# now let's look at splitting the object-, inside groupby paas the list of columns which we want group


country_grp = dataset.groupby(['Country'])
country_grp
```

```python
# so here DataFrameGroupBy contains bunch of groups, now let's take the look at individual group which it has
```

```python
country_grp.get_group('India')

# we can see that all the dataframes are based on country filter
```

```python
# so far nothing special this can be achieved by filter for a single column-

filt = dataset['Country']=='United States'
dataset.loc[filt].head(5)
```

```python
# Now lets filter the socialmedia column based on the country-

filt = dataset['Country']=='United States'
# Here above in the filter condition anytime we can change the country name to any

dataset.loc[filt]['SocialMedia'].value_counts()
```

#### Calculate the results in percentage .


```python
# I'll calculate the total number of respondent from each country who participated in survey 

country_respondent = dataset['Country'].value_counts()
country_respondent
```

```python
# Now I'll get the total number of people from each country who know python 

people_knows_python = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
people_knows_python
```

```python
# So we have two series now I'll combine these two 

python_knows_df = pd.concat([country_respondent,people_knows_python], axis='columns', sort=False)
python_knows_df
```

```python
# Let's rename the column for better understanding

python_knows_df.rename(columns={'Country':'Num_Respondent','LanguageWorkedWith':'Num_Knows_Python' }, inplace=True)
python_knows_df
```

```python
# Now we have total number of respondent from each country and the number of people who know python in a single data frame
# Calculate a percentage

python_knows_df['Pct_Knows_python'] = (python_knows_df['Num_Knows_Python']/python_knows_df['Num_Respondent'])*100
python_knows_df
```

```python
# Now let's sort the country by largest percentage

python_knows_df.sort_values(by='Pct_Knows_python', ascending=False, inplace=True)
python_knows_df.head(50)
```

```python
# Now I can get the results for a specific contry

python_knows_df.loc['India']
```
