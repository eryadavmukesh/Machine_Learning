```python
persons = {
    'first_name':['Mukesh','Emma','Ashish','Willie','joe'],
    'last_name':['Yadav', 'Watson','Agarwal','meeks','knack'],
    'occupation':['Engineer','Product Ownor','Architect','Infra Admin','Vendor mgmt'],
    'location':['Jaipur', 'Delhi', 'Pink city','USA', 'Canada'],
    'pinCode':['303106', '302020', '302016','12321','344354'],
    'hobby':['Reading', 'Study','playing','Hockey','Baseball'],
    'email':['mukesh@gmail.com','emma@hotmail.com','ashish@yahoo.com','willie@mymail.com','joe@rediffmail.com']
        }
```

```python
import pandas as pd
df = pd.DataFrame(persons)
df
```

```python
#Sort based on descending order

df.sort_values(by='last_name', ascending=False)
```

```python
# Sometime if some columns can have identical values and you want to sort based on multiple columns -

df.sort_values(['last_name', 'first_name'], ascending=False)
```

```python
# Sometime you want to sort based on multiple columns with different conditions -

df.sort_values(['last_name','first_name'], ascending=[False, True])
```

```python
df.sort_values(['last_name','first_name'], ascending=[False, True], inplace=True)
```

```python
df
```

```python
# If you want to make the changes back and index should be earlier one - 

df.sort_index()
```

```python
# If you want to sort the series instead of entire data then - 

df['last_name'].sort_values()
```

```python
data = pd.read_csv('developer_survey_2019/survey_results_public.csv')
```

```python
data.head(5)
```

```python
# so if you want to sort this data based on country column-
data.sort_values(by='Country', inplace=True)
data['Country'].head(60)
```

```python

data[['Country','ConvertedComp']].head(40)
```

```python
# So now if you want to sort salary by descending order and Country by asc then-

data.sort_values(['Country','ConvertedComp'], ascending=[True, False], inplace=True)
data[['Country','ConvertedComp']].head(40)

# Here we can see outliers are there 
```

```python
# Now let's grab the largest and smallest values from the data 

#let's grab the 15 largest values from the ConvertedComp columns

data['ConvertedComp'].nlargest(15)
```

```python
# Ther's other way to run nlargest on entire data 

data.nlargest(10, 'ConvertedComp')
```

```python
# so just to get the smallest values-

data.nsmallest(10, 'ConvertedComp').head(10)
```
