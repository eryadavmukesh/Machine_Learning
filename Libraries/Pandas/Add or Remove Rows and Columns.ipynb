```python
persons = {
    'FName':['Mukesh','Admin','Raja','Willie','joe'],
    'LName':['Yadav', 'admin','Sastri','Meeks','knack'],
    'Occupation':['Engineer','Product Ownor','Architect','Infra Admin','Vendor mgmt'],
    'Location':['Jaipur', 'Delhi', 'Pink city','USA', 'Canada'],
    'PinCode':['303106', '302020', '302016','12321','344354'],
    'Hobby':['Reading', 'Study','playing','Hockey','Baseball']
             }
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(persons)
```

```python
df
```

### Add/ Remove columns

```python
# Now let's add a column

df['FName']+ ' '+ df['LName']
```

```python
df
```

```python
df['FullName'] = df['FName']+ ' '+ df['LName']
```

```python
df
```

```python
# Now let's remove a column
df.drop(columns=['FName','LName'])
```

```python
df
```

```python
df.drop(columns=['FName','LName'], inplace=True)
```

```python
df
```

```python
#Now let's do a reverse process to split FullName column using split from string
df['FullName'].str.split(' ')
```

```python
# so now if we want to have two seperate columns then we have to do expand - 
df['FullName'].str.split(' ', expand=True)
```

```python
df[['first', 'last']] = df['FullName'].str.split(' ', expand=True)
```
